```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

data = pd.read_csv("C:/Users/Maruata/Desktop/dcnns/dcnns.csv")
```

```python
history_dict = data

dcnn_lr_001_accuracy = history_dict['dcnn_lr_001_accuracy']
dcnn_lr_001_val_accuracy = history_dict['dcnn_lr_001_val_accuracy']
dcnn_lr_0001_accuracy = history_dict['dcnn_lr_0001_accuracy']
dcnn_lr_0001_val_accuracy = history_dict['dcnn_lr_0001_val_accuracy']

start_by_epoch = 1
epochs = range(start_by_epoch, len(dcnn_lr_001_accuracy) + 1)

sn.set(font_scale=2.5)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")

plt.plot(epochs, dcnn_lr_001_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_001_val_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_0001_accuracy[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_0001_val_accuracy[start_by_epoch-1:], '-', linewidth=2)

plt.title('Comparison of Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.legend(['Accuracy with 0.001', 'Validation Accuracy with 0.001', 
           'Accuracy with 0.0001', 'Validation Accuracy with 0.0001'            
           ], loc='center right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
#plt.ylim(0.9565,1.002)

plt.show()

fig.savefig('accuracy.pdf') 

plt.clf()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")

dcnn_lr_001_loss = history_dict['dcnn_lr_001_loss']
dcnn_lr_001_val_loss = history_dict['dcnn_lr_001_val_loss']
dcnn_lr_0001_loss = history_dict['dcnn_lr_0001_loss']
dcnn_lr_0001_val_loss = history_dict['dcnn_lr_0001_val_loss']

plt.plot(epochs, dcnn_lr_001_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_001_val_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_0001_loss[start_by_epoch-1:], '-', linewidth=2)
plt.plot(epochs, dcnn_lr_0001_val_loss[start_by_epoch-1:], '-', linewidth=2)


plt.title('Comparison of Training Loss with lr = 0.001')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.legend(['Loss with 0.001', 'Validation Loss with 0.001', 
           'Loss with 0.0001', 'Validation Loss with 0.0001'            
           ], loc='upper right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)

plt.tight_layout()
plt.show()

fig.savefig('loss.pdf') 

plt.clf()
```
